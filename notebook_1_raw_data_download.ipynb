STEP 2 - IMPORT LIBRARIES

```python
from datasets import load_dataset
import os
from google.colab import drive
drive.mount('/content/drive')
```

STEP 3 - PROJECT PATH

```python
PROJECT_PATH = (
    "/content/drive/MyDrive/"
    "RetailBot_AI_Powered_ECommerce_Customer_Service_LLM"
)
RAW_DATA_PATH = f"{PROJECT_PATH}/data/raw"

os.makedirs(RAW_DATA_PATH, exist_ok=True)

print("Project Path:")
print(PROJECT_PATH)

print("\nRaw Data Path:")
print(RAW_DATA_PATH)
```

STEP 4 - DOWNLOAD DATASETS
Bitext Customer Support

```python
# Bitext Customer Support
print("Loading Bitext Customer Support Dataset...")

bitext = load_dataset(
    "bitext/Bitext-customer-support-llm-chatbot-training-dataset"
)

print(bitext)
```

```python
# Customer Service Dialogues
print("Loading Customer Service Dialogues Dataset...")

dialogues = load_dataset(
    "FunDialogues/customer-service-grocery-cashier"
)

print(dialogues)
```

```python
# E-Commerce Support
print("Loading E-Commerce Support Dataset...")

ecommerce = load_dataset(
    "NebulaByte/E-Commerce_Customer_Support_Conversations"
)

print(ecommerce)
```

```python
# Ecommerce FAQ
print("Loading Ecommerce FAQ Dataset...")

faq = load_dataset(
    "dltdojo/ecommerce-faq-chatbot-dataset"
)

print(faq)
```

```python
# STEP 5 - VERIFY DATASET STRUCTURE
print("\n==============================")
print("BITEXT DATASET")
print("==============================")

print(bitext["train"].column_names)

print("\nSample Record:")
print(bitext["train"][0])
```

```python
print("\n==============================")
print("CUSTOMER SERVICE DIALOGUES")
print("==============================")

print(dialogues["train"].column_names)

print("\nSample Record:")
print(dialogues["train"][0])
```

```python
print("\n==============================")
print("ECOMMERCE SUPPORT")
print("==============================")

print(ecommerce["train"].column_names)

print("\nSample Record:")
print(ecommerce["train"][0])
```

```python
print("\n==============================")
print("ECOMMERCE FAQ")
print("==============================")

print(faq["train"].column_names)

print("\nSample Record:")
print(faq["train"][0])
```

```python
# STEP 6 - SAVE RAW DATASETS

bitext["train"].save_to_disk(
    f"{RAW_DATA_PATH}/bitext_customer_support"
)

dialogues["train"].save_to_disk(
    f"{RAW_DATA_PATH}/customer_service_dialogues"
)

ecommerce["train"].save_to_disk(
    f"{RAW_DATA_PATH}/ecommerce_support"
)

faq["train"].save_to_disk(
    f"{RAW_DATA_PATH}/ecommerce_faq"
)
```

```python
# STEP 7 - VERIFY SAVED DATA
import os

print("\nSaved Datasets:")

for folder in os.listdir(RAW_DATA_PATH):
    print(folder)
```

```python
# STEP 8 - FINAL SUMMARY
print("=" * 60)

print("RetailBot Data Collection Completed")

print("=" * 60)

print("\nDatasets Stored At:")

print(RAW_DATA_PATH)

print("\nReady For:")

print("Notebook 2 - Data Cleaning & Preprocessing")
```
